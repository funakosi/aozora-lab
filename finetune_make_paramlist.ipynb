```python
# ハイパーパラメータのリスト
type_list = ['constant',] #'linear', 'cosine'] # スケジューリングのタイプ
lr_list = [5e-6, 7e-6, 1e-5, 1.5e-5, 2e-5] # 学習率
batch_size_list = [16, 32, 64, 128, 256, 512] # バッチサイズ
weight_decay_list = [0, 1e-4, 5e-4, 1e-3] # 正則化
epoch_list = [2] # epoch(Loop内で更新していく)
```

```python
import itertools
import pprint
hyper_parameters = []
for i, p in enumerate(itertools.product(type_list, lr_list, batch_size_list, weight_decay_list)):
    if p[2] >= 512:
        batch_size = 128
        acc_steps = p[2] / batch_size
    else:
        batch_size = p[2]
        acc_steps = 1
    para = {
        'epoch': 1,
        'type': p[0],
        'lr': p[1],
        'batch': batch_size,
        'weight_decay': p[3],
        'accumulation_steps': acc_steps,
        'target': True,
    }
    hyper_parameters.append(para)
print(len(hyper_parameters))
# pprint.pprint(hyper_parameters, width=130)
```

```python
import pandas as pd
df = pd.DataFrame(hyper_parameters)
df.to_csv('data/hyper_parameters.csv')
```
